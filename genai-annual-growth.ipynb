```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
genai_total_market_csv = 'genai_total_market.csv'
```

```python
genai_total_market_df = pd.read_csv(genai_total_market_csv)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax = sns.barplot(data=genai_total_market_df, x='Year', y='Revenues', color = 'blue')

ax.bar_label(ax.containers[0], fmt='%.0f B$', size=14, label_type='center', color='white')

plt.title('Gen AI Estimated Market Size (Bloomberg Intelligence)', size = 24)

plt.xlabel("Years",    size = 18)
plt.ylabel("Revenues", size = 18)

plt.savefig("genai_total_market.png")
```

```python
genai_relative_spend_csv = 'genai_relative_spend.csv'
```

```python
genai_relative_spend_df = pd.read_csv(genai_relative_spend_csv)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

ax = sns.barplot(data=genai_relative_spend_df, x='Year', y='Percentage', color = 'green')

ax.bar_label(ax.containers[0], fmt='%.0f %%', size=14, label_type='center', color='white')

plt.title('Gen AI vs. Total Technology Spend (Bloomberg Intelligence)', size = 24)

plt.xlabel("Years",      size = 18)
plt.ylabel("Percentage", size = 18)

plt.savefig("genai_relative_spend.png")
```
